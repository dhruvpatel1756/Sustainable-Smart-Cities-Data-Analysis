```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (8,5)
```

```python
#load dataset 
df = pd.read_csv('smart_city_citizen_activity.csv')
df.head()
```

# Visualization

```python
sns.countplot(x='Gender',data=df)
plt.title('Gender Distribution')
plt.show()
```

### Female has more count then Male.

```python
sns.countplot(x='Mode_of_Transport',data=df)
plt.xticks(rotation=45)
plt.title('Mode of Transport Distribution')
plt.show()
```

```python
sns.boxplot(x='Mode_of_Transport',y='Carbon_Footprint_kgCO2',data=df)
plt.xticks(rotation=45)
plt.title('Transport Mode vs Carbon Footprint')
plt.show()
```

```python
sns.histplot(df['Home_Energy_Consumption_kWh'],bins=20,kde=True)
plt.title('Home Energy Consumption Distribution')
plt.show()
```

```python
sns.scatterplot(x="Home_Energy_Consumption_kWh",y="Carbon_Footprint_kgCO2",data=df)

plt.title("Energy Consumption vs Carbon Footprint")
plt.show()

print("Correlation:",df["Home_Energy_Consumption_kWh"].corr(df["Carbon_Footprint_kgCO2"]))
```

```python
sns.scatterplot(x="Steps_Walked",y="Calories_Burned",data=df)

plt.title("Steps vs Calories Burned")
plt.show()
```

### More steps walked more calories burned.

```python
sns.scatterplot(x="Work_Hours",y="Sleep_Hours",data=df)

plt.title("Work Hours vs Sleep Hours")
plt.show()
```

```python
sns.histplot(df["Charging_Station_Usage"], bins=15, kde=True)
plt.title("EV Charging Station Usage")
plt.show()
```

```python
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
high_carbon = df[df["Carbon_Footprint_kgCO2"] > df["Carbon_Footprint_kgCO2"].mean()]

print("High Carbon Footprint Citizens:")
print(high_carbon[["Citizen_ID", "Mode_of_Transport","Home_Energy_Consumption_kWh","Carbon_Footprint_kgCO2"]].head())
```

### The 5 Person has High Carbon Footprints.
